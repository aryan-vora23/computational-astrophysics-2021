```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Astrosat_Catalog.csv")
df.loc[df['ra']>180, 'ra']=df['ra']-360
```

```python
lmxb_df=df[(df['Final_Type']=='LMXB')]
```

```python
hmxb_df=df[(df['Final_Type']=='HMXB')]
```

```python
fig=plt.figure(figsize=(20,20))
ax1=fig.add_subplot(121,projection="mollweide",facecolor='black')
ax1.plot(np.deg2rad(lmxb_df['ra']),np.deg2rad(lmxb_df['dec']),',',c='#7CFC00')
ax1.set(title='LMXB')
ax1.grid(c='w')
for label in ax1.get_xticklabels()+ax1.get_yticklabels():
    label.set_bbox(dict(facecolor='white'))
ax2=fig.add_subplot(122,projection="mollweide",facecolor='black')
ax2.plot(np.deg2rad(hmxb_df['ra']),np.deg2rad(hmxb_df['dec']),',',c='#7CFC00')
ax2.set(title='HMXB')
ax2.grid(c='w')
for label in ax2.get_xticklabels()+ax2.get_yticklabels():
    label.set_bbox(dict(facecolor='white'))
plt.show()
#The arc is representing the Milky Way galaxy in the HMXB subplot
```

```python
f1_df=df[(df['Astrosat_Flag']==1)]
f0_df=df[(df['Astrosat_Flag']==0)]
```

```python
fig2=plt.figure(figsize=(20,20))
x1=fig2.add_subplot(121,projection="mollweide",facecolor='black')
x1.plot(np.deg2rad(f1_df['ra']),np.deg2rad(f1_df['dec']),',',c='#7CFC00')
x1.set(title='Sources observed by AstroSAT')
x1.grid(c='w')
for label in x1.get_xticklabels()+ax1.get_yticklabels():
    label.set_bbox(dict(facecolor='white'))
x2=fig2.add_subplot(122,projection="mollweide",facecolor='black')
x2.plot(np.deg2rad(f0_df['ra']),np.deg2rad(f0_df['dec']),',',c='#7CFC00')
x2.set(title='Sources not observed by AstroSAT')
x2.grid(c='w')
for label in x2.get_xticklabels()+ax2.get_yticklabels():
    label.set_bbox(dict(facecolor='white'))
plt.show()
```

```python
g0_df=lmxb_df[(lmxb_df['Astrosat_Flag']==0)]
g1_df=lmxb_df[(lmxb_df['Astrosat_Flag']==1)]
g2_df=hmxb_df[(hmxb_df['Astrosat_Flag']==0)]
g3_df=hmxb_df[(hmxb_df['Astrosat_Flag']==1)]
```

```python
fig3=plt.figure(figsize=(20,20))
pfinal=fig3.add_subplot(projection="mollweide",facecolor='black')
pfinal.set(title='All Stars')
plt.plot(np.deg2rad(g0_df['ra']),np.deg2rad(g0_df['dec']),',',c='#FFD700',label='LMXB not observed by AstroSAT')
plt.plot(np.deg2rad(g1_df['ra']),np.deg2rad(g1_df['dec']),',',c='#00BFFF',label='LMXB observed by AstroSAT')
plt.plot(np.deg2rad(g2_df['ra']),np.deg2rad(g2_df['dec']),',',c='#7CFC00',label='HMXB not observed by AstroSAT')
plt.plot(np.deg2rad(g3_df['ra']),np.deg2rad(g3_df['dec']),',',c='#FF00FF',label='HMXB observed by AstroSAT')
for label in pfinal.get_xticklabels()+ax1.get_yticklabels():
    label.set_bbox(dict(facecolor='white'))
leg=pfinal.legend(fancybox=True,facecolor='black',loc='upper right')
for text in leg.get_texts():
    text.set_color("white")
plt.grid(True)
plt.show()
```
